## Decision Tree (Chris Mariani)

```python
from heading_v1 import *
from utils import *
from sklearn.tree import DecisionTreeRegressor 
from sklearn import metrics
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
from matplotlib import pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt

sales = DataSet(name="VideoGameSalesUpdated", target='Global_Sales', attr_names='NA_Sales  JP_Sales  Global_Sales')

# Training Test Datasets
train_data = []
train_label = []
test_data = []
test_label = []

train_data, train_label, test_data, test_label = seperate_train_and_test_data(sales.examples, 100)

futureSample_data =[]
futureSample_label = []
futureSample_data= test_data[-2:]
futureSample_label= test_label[-2:]

test_data = test_data[:-2]
test_label = test_label[:-2]

print(futureSample_data[:], futureSample_label[:])

# Decision Tree Implementation
model = DecisionTreeRegressor()
model.fit(train_data, train_label)

# Model Testing
solution_test = model.predict(test_data)

y_test = test_label
pred = solution_test

print("r^2 = " + str(metrics.r2_score(y_test, pred)))


plt.scatter(y_test, pred)
plt.xlabel('y_test')
plt.ylabel('pred')
plt.show()

## Prediction and for Future Sample Set
solution_validate = model.predict(futureSample_data)

for i in range (2):
    print("The predicted value is {} and the actual value is {}".format(solution_validate[i],futureSample_label[i]))

```
