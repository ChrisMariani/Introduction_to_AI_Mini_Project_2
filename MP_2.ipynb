## K Nearest Regression (Dylan Nadan)

```python
from heading_v1 import *
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
#_______________________________________
sales = DataSet(name="VideoGameSalesUpdated", target='Global_Sales', attr_names='NA_Sales JP_Sales Global_Sales')
#_______________________________________
train_data = []
train_label = []
test_data = []
test_label = []
#_______________________________________
train_data, train_label, test_data, test_label = seperate_train_and_test_data(sales.examples, 3320)
#_______________________________________
futureSample_data =[]
futureSample_label = []
futureSample_data= test_data[-2:]
futureSample_label= test_label[-2:]
#_______________________________________
test_data = test_data[:-2]
test_label = test_label[:-2]
#_______________________________________

#_______________________________________
 
model = KNeighborsRegressor(n_neighbors=129)
model.fit(train_data, train_label)
#_______________________________________
```

```python
from sklearn import metrics
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
from matplotlib import pyplot as plt
import pandas as pd
#_______________________________________
solution_test = model.predict(test_data)
#_______________________________________
y_test = test_label
pred = solution_test
#_______________________________________
print("R^2 = " + str(metrics.r2_score(y_test, pred)))
#_______________________________________
print("Mean Absolute Error = " + str(metrics.mean_absolute_error(y_test, pred)))
#_______________________________________

print("Mean-squared Error = " + str(metrics.mean_squared_error(y_test, pred)**0.5))
#_______________________________________
plt.scatter(y_test, pred, c = y_test, edgecolor = "k")
plt.xlabel('y_test')
plt.ylabel('pred')
plt.show()
#_______________________________________
solution_validate = model.predict(futureSample_data)
#_______________________________________
for i in range (2):
    print("The predicted value is {} and the actual value is {}".format(solution_validate[i],futureSample_label[i]))
#_______________________________________
NA_sales = input("NA sales: ")
#_______________________________________
JP_sales = input("JPN sales: ")
input_sale = []
input_sale.append(NA_sales)
input_sale.append(JP_sales)

input_data = []
input_data.append(input_sale)

output_model = model.predict(input_data)

print("Global Sales: " + str(output_model[0]))
#_______________________________________
```
